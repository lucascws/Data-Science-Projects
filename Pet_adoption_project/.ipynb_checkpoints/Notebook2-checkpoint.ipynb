## Text Embeddings

```python
from sentence_transformers import SentenceTransformer
```

In this section, the purpose is to extract text features from the Profile description write-up to input as features for our model. To achieve this, I would need to get features in the form of a vector representation for the text. In order to use them as input for the model, I would need to ensure that the output for each description is a fixed vector of the same size. 

One way to do this is to use a Count Vectorizer or TF-IDF. However, as seen from our EDA section, the top words for pets that get adopted and pets that do not get adopted are largely the same. This method would not be effective. 

Another way we can get the embeddings for the sentence is to get the word embeddings, then average the word embeddings for the words in the document. We can do this using libraries like Word2Vec or GloVe. However, the limitations for this is that this method does not take into account the order or the context of the words when generating the embeddings. This means that documents containing the same words in different order will have the exact same vector representation. Another issue is that some words can have the different meanings when used in different sentences. 

To ensure that the contextual meaning of the sentence is retained when generating the embeddings, I will use a BERT (Bidirectional Encoder Representations) model which can overcome the limitations of the embeddings generated from Word2Vec/GloVe. They take into consideration the context and can differentiate words with multiple meaning using its bidirectional capability (other models only read text input sequentially). Specifically for this case, I will be using a Sentence BERT (SBERT) model, which is a modification of the BERT model, using a siamese and triplet network structures. The SBERT model is used in order to obtain a fixed size text embedding for each description. 

[reference](https://medium.com/@kashyapkathrani/all-about-embeddings-829c8ff0bf5b)

This pre-trained Sentence BERT model (all-MiniLM-L6-v2) was chosen as it was a good balance of speed and embeddings quality. It outputs each description into a vector with 384 elements.
[source](https://www.sbert.net/docs/pretrained_models.html)

```python
text_df = pd.read_csv('data/text_df.csv')
```

```python
text_df.head()
```

```python
# Import pre-trained model

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
# get embeddings for each description and store them in a dictionary

text_embeddings_dict = {}
for pet_id, text in tqdm_notebook(zip(text_df['PetID'], text_df['Description'])):
    text_embeddings_dict[pet_id] = model.encode(text)
```

```python
# function to create dataframe from embedding dictionary
def create_embeddings_df(embeddings_dict, col_name, key):
    embeddings_df = pd.DataFrame.from_dict(embeddings_dict, orient='index')
    new_col_name = []
    for col in embeddings_df.columns:
        new_col_name.append(col_name + '_' + str(col))
    embeddings_df.columns = new_col_name
    embeddings_df = embeddings_df.reset_index()
    embeddings_df = embeddings_df.rename({'index': key}, axis=1)
    return embeddings_df
```

```python
text_embeddings = create_embeddings_df(text_embeddings_dict, 'text', 'PetID')
```

## Image Embeddings

```python
import cv2
import os

from tensorflow.keras.applications.efficientnet_v2 import preprocess_input, EfficientNetV2B3
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import AveragePooling1D, Lambda, Input, MaxPooling2D, GlobalMaxPooling2D
import tensorflow.keras.backend as K
```

For this section, the goal is to extract features from the profile picture of each pet in the form of image embeddings. To obtain the embeddings for the images, I will be using a CNN model. Convolutional Neural Network (CNN) are a class of Neural network that are specialized in processing image data. They generally consist of the convolution and pooling layer, which allows us to get a vector representation of the iamge in a much smaller size.

For this case, I will be using the pretrained EfficientNet model as it provides a good balance and embeddings quality. The output for the pretrained model is a vector with the dimension (10, 10, 1536). To reduce the vector into a size to fit into the model, I will perform Max Pooling and Average pooling, getting a vector size suitable for modelling. I will resize the image with paddings to retain the original length and width ratio of the photo. 

[source](https://data-science-blog.com/blog/2022/04/11/how-to-choose-the-best-pre-trained-model-for-your-convolutional-neural-network/)

```python
# Read image data
image_df = pd.read_csv('data/image_df.csv')
```

```python
image_df.info()
```

```python
image_df.head()
```

```python
# define parameters for the below function
batch_size = 16
pet_profile_pics = image_df['ProfilePic'].values
n_batches = len(pet_profile_pics) // batch_size + 1
```

I will be resizing the images for the pretrained model. To retain the original dimensions, I will include paddings at the sides/top/bottom to avoid contorting the image 

```python
# function for processing and loading image to the correct size for pre-trained model 
img_size=300
def resize_image(im):
    # Get original size of image in height and width
    original_size = im.shape[:2]
    ratio = float(img_size)/max(original_size)
    # resize both length and width with the same ratio so image will not be distorted
    new_size = tuple([int(measurements*ratio) for measurements in original_size])
    im = cv2.resize(im, (new_size[1], new_size[0]))
    # find empty spaces area the shorter size
    delta_w = img_size - new_size[1]
    delta_h = img_size - new_size[0]
    top, bottom = delta_h//2, delta_h-(delta_h//2)
    left, right = delta_w//2, delta_w-(delta_w//2)
    # padding with black color
    color = [0, 0, 0]
    # copy image into the middle and fill sides/top/bottom empty spaces with black color
    new_im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT,value=color)
    return new_im

def load_image(pet_profile_pic, path='pet_images/'):
    image = cv2.imread(f'{path}{pet_profile_pic}')
    new_image = resize_image(image)
    new_image = preprocess_input(new_image)
    return new_image
```

```python
# get pretrained model
inp = Input((img_size,img_size,3))
pre_trained_model = EfficientNetV2B3(input_tensor = inp, include_top = False)
pre_trained_model.trainable = False
```

I will do a average pooling from the output from the model to reduce the size of the vector embeddings for modelling

```python
# build layers to extract embeddings

feature_extractor = Sequential()

feature_extractor.add(pre_trained_model)

# to change vector to 1D
feature_extractor.add(GlobalMaxPooling2D())

# change dimension to perform average pooling to reduce size of output
feature_extractor.add(Lambda(lambda x: K.expand_dims(x,axis = -1)))

# reduce size of vector by average pooling
feature_extractor.add(AveragePooling1D(4))

# change dimension to get 1d vector as output
feature_extractor.add(Lambda(lambda x: x[:,:,0]))
```

```python
feature_extractor.summary()
```

```python
# create model for image feature extraction

outputs = feature_extractor.get_layer('lambda_1').output
model = Model(inputs = feature_extractor.inputs, outputs= outputs)
```

```python
model.summary()
```

Next, I will get the image embeddings and store them in a dictionary.

```python
features = {}
for batch in tqdm_notebook(range(n_batches)):
    start = batch*batch_size
    end = (batch+1)*batch_size
    batch_pets_pics = pet_profile_pics[start:end]
    batch_images = np.zeros((len(batch_pets_pics),img_size,img_size,3))
    
    for i, pet_pic in enumerate(batch_pets_pics):
        try:
            batch_images[i] = load_image(pet_pic)
        except:
            pass
    batch_preds = model.predict(batch_images, verbose=0)
    for i,pet_pic in enumerate(batch_pets_pics):
        features[pet_pic] = batch_preds[i];
```

```python
# convert embeddings into a dataframe
image_embeddings = create_embeddings_df(features, 'image', 'ProfilePic')
```

```python
# save model
model.save('models/image_embeddings.h5')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

In this section, I will perform the same steps to retrieve the image embedding but I will skip the average pooling here for the image search to keep more information. Using embeddings, I will then perform the image search.

```python
# only keep profiles with a profile picture
image_search_df = image_df[image_df['PhotoAmt'] != 0]
```

```python
# create model for feature extraction

image_extractor = Sequential()
image_extractor.add(pre_trained_model)
image_extractor.add(GlobalMaxPooling2D())
```

```python
# define layer to get output

image_outputs = image_extractor.get_layer('global_max_pooling2d_1').output
image_model = Model(inputs = image_extractor.inputs, outputs= image_outputs)
```

```python
image_model.summary()
```

```python
# parameters for feature extraction

batch_size = 16
pet_profile_pics = image_search_df['ProfilePic'].values
n_batches = len(pet_profile_pics) // batch_size + 1
```

```python
# Extract features

features = {}
for batch in tqdm_notebook(range(n_batches)):
    start = batch*batch_size
    end = (batch+1)*batch_size
    batch_pets_pics = pet_profile_pics[start:end]
    batch_images = np.zeros((len(batch_pets_pics),img_size,img_size,3))
    
    for i, pet_pic in enumerate(batch_pets_pics):
        batch_images[i] = load_image(pet_pic)

    batch_preds = model.predict(batch_images, verbose=0)
    for i,pet_pic in enumerate(batch_pets_pics):
        features[pet_pic] = batch_preds[i]
```

```python
# embeddings in dataframe

image_embeddings = pd.Series(features).reset_index()
image_embeddings = image_embeddings.rename({'index': 'img_path', 0:'embeddings'}, axis=1)
```

```python
# get image file path in img_path column

image_embeddings['img_path'] = image_embeddings['img_path'].apply(lambda x: "pet_images/" + x)
```

In the next portion, I will perform an image search using the cosine similarity and KNN

1. Randomly select a pet from listing
2. Compute the cosine similarity score of the selected pet against all the other pets
3. Sort the pets by the similarity score in descending order to get the most similar looking pets
4. Determine the number of similar pets using KNN (K-Nearest Neighbor) to find - k=5, show the top 5 closest pets


```python
# Function to plot similar looking pets based on cosine similarity

def plot_similar(df, embedding_col, index, quantity=5):
   
    # Calculate pairwise cosine similarities between query and all rows
    similarities = cosine_similarity([df[embedding_col][index]], df[embedding_col].values.tolist())[0]
    
    # Find most similar image index
    k = quantity + 1
    similarities_series = pd.Series(similarities)
    nearest_index = similarities_series.sort_values(ascending=False).head(k).index
    nearest_index = list(nearest_index[nearest_index != index])
    
    # Plot input image
    img = cv2.imread(df['img_path'][index])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # img = Image.open(df['img_path'][query_index]).convert('RGB')
    plt.imshow(img)
    plt.grid(False)
    plt.axis('off')
    plt.title('Selected Pet')
    
    # Plot nearest neighbors images
    fig = plt.figure(figsize=(20,4))
    plt.suptitle('Similar Pets')
    for idx, neighbor in enumerate(nearest_index):
        plt.subplot(1, len(nearest_index), idx+1)
        img = cv2.imread(df['img_path'][neighbor])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
        plt.grid(False)
        plt.axis('off')
        plt.title(f'Cosine Sim: {similarities[neighbor]:.3f}')
    plt.tight_layout()
```

```python
plot_similar(image_embeddings, 'embeddings', 3951)
```

```python
plot_similar(image_embeddings, 'embeddings', 50)
```

Limitations:

The limitation of this method is its scalability. To perform image search using KNN, we have to compute the similarity score against all other items in the dataset. When the dataset gets larger, this process will take too long. Using an Aproximate Nearest Neighbour (ANN) model would be helpful for this case.
